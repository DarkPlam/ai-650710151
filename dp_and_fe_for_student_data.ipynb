##ปรับชื่อคณะให้เป็นมาตรฐานเหมือนกัน เช่น รวมชื่อย่อและภาษาอังกฤษให้เป็นชื่อเดียวกัน
*ยกตัวอย่าง ('วิทยาศาสตร์', 'วิทย์', 'Science', 'คณะวิทยาศาสตร์') เปลี่ยนให้เป็นคำเดียวกัน คือ วิทยาศาสตร์

```python
def standardize_faculty(name):
    if name in ['วิทยาศาสตร์', 'วิทย์', 'Science', 'คณะวิทยาศาสตร์']:
        return 'วิทยาศาสตร์'
    elif name in ['ศิลปศาสตร์', 'ศิลป์', 'Arts']:
        return 'ศิลปศาสตร์'
    else:
        return name
```

##เราจะสร้างคอลัมน์ใหม่ ปรับชื่อคณะให้เป็นมาตรฐาน

```python
df['คณะ_มาตรฐาน'] = df['คณะ'].apply(standardize_faculty)
```

##คำนวณควอไทล์ที่ 25% (ค่าต่ำสุดของกลุ่ม)
คือ**Q1 = df['อายุ'].quantile(0.25)**
## คำนวณควอไทล์ที่ 75% (ค่าสูงสุดของกลุ่ม)
คือ**Q3 = df['อายุ'].quantile(0.75)**


```python
Q1 = df['อายุ'].quantile(0.25)
Q3 = df['อายุ'].quantile(0.75)
```

##แสดงค่าควอไทล์ที่ 25% และ 75% ของคอลัมน์ 'อายุ'
ผลลัพธ์  **Q1** คือ 20.0 **Q2** คือ 22.0

```python
Q1, Q3
```

##หาช่วงควอไทล์ (Interquartile Range) ความต่างระหว่าง Q3 และ Q1
*หาได้จาก Q3 - Q1

```python
IQR = Q3 - Q1
```

##แสดงค่าช่วงควอไทล์ (Interquartile Range) ของอายุ
*ผลลัพธ์จะได้ 2.0 คือค่าช่วงควอไทล์ห่างกันแค่ 2 ช่วง

```python
IQR
```

##ค่าขอบล่างสำหรับตรวจจับค่าผิดปกติ (outlier)
*lower_bound = Q1 - 1.5 * IQR  
##ค่าขอบบนสำหรับตรวจจับค่าผิดปกติ (outlier)
*upper_bound = Q3 + 1.5 * IQR

```python
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
```

##ให้แสดงผลของ ค่าขอบล่างและค่าขอบบนของอายุ

```python
print(lower_bound)
print(upper_bound)
```

##แสดงข้อมูล DAtaFrame ที่กรองข้อมูลเฉพาะแถวที่อายุอยู่ในช่วงค่าปกติ

```python
df_no_outliers = df[(df['อายุ'] >= lower_bound) & (df['อายุ'] <= upper_bound)]
```

##ให้แสดงข้อมูล DataFrame ที่กรองเอาเฉพาะแถวที่ไม่มีค่าอายุผิดปกติ

```python
df_no_outliers
```

##ให้ตรวจสอบว่าแต่ละแถวมีค่าอายุมากกว่าหรือเท่ากับขอบล่าง (ผลลัพธ์เป็น True/False)

```python
df['อายุ'] >= lower_bound
```

##ให้ตรวจสอบว่าแต่ละแถวมีค่าอายุน้อยกว่าหรือเท่ากับขอบบน (ผลลัพธ์เป็น True/False)

```python
df['อายุ'] <= upper_bound
```
