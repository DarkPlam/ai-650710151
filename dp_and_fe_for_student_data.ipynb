##‡∏Ñ‡∏≥‡∏ô‡∏ß‡∏ì‡∏Ñ‡∏ß‡∏≠‡πÑ‡∏ó‡∏•‡πå‡∏ó‡∏µ‡πà 25% (‡∏Ñ‡πà‡∏≤‡∏ï‡πà‡∏≥‡∏™‡∏∏‡∏î‡∏Ç‡∏≠‡∏á‡∏Å‡∏•‡∏∏‡πà‡∏°)
‡∏Ñ‡∏∑‡∏≠**Q1 = df['‡∏≠‡∏≤‡∏¢‡∏∏'].quantile(0.25)**
## ‡∏Ñ‡∏≥‡∏ô‡∏ß‡∏ì‡∏Ñ‡∏ß‡∏≠‡πÑ‡∏ó‡∏•‡πå‡∏ó‡∏µ‡πà 75% (‡∏Ñ‡πà‡∏≤‡∏™‡∏π‡∏á‡∏™‡∏∏‡∏î‡∏Ç‡∏≠‡∏á‡∏Å‡∏•‡∏∏‡πà‡∏°)
‡∏Ñ‡∏∑‡∏≠**Q3 = df['‡∏≠‡∏≤‡∏¢‡∏∏'].quantile(0.75)**


```python
Q1 = df['‡∏≠‡∏≤‡∏¢‡∏∏'].quantile(0.25)
Q3 = df['‡∏≠‡∏≤‡∏¢‡∏∏'].quantile(0.75)
```

##‡πÅ‡∏™‡∏î‡∏á‡∏Ñ‡πà‡∏≤‡∏Ñ‡∏ß‡∏≠‡πÑ‡∏ó‡∏•‡πå‡∏ó‡∏µ‡πà 25% ‡πÅ‡∏•‡∏∞ 75% ‡∏Ç‡∏≠‡∏á‡∏Ñ‡∏≠‡∏•‡∏±‡∏°‡∏ô‡πå '‡∏≠‡∏≤‡∏¢‡∏∏'
‡∏ú‡∏•‡∏•‡∏±‡∏û‡∏ò‡πå  **Q1** ‡∏Ñ‡∏∑‡∏≠ 20.0 **Q2** ‡∏Ñ‡∏∑‡∏≠ 22.0

```python
Q1, Q3
```

##‡∏´‡∏≤‡∏ä‡πà‡∏ß‡∏á‡∏Ñ‡∏ß‡∏≠‡πÑ‡∏ó‡∏•‡πå (Interquartile Range) ‡∏Ñ‡∏ß‡∏≤‡∏°‡∏ï‡πà‡∏≤‡∏á‡∏£‡∏∞‡∏´‡∏ß‡πà‡∏≤‡∏á Q3 ‡πÅ‡∏•‡∏∞ Q1
*‡∏´‡∏≤‡πÑ‡∏î‡πâ‡∏à‡∏≤‡∏Å Q3 - Q1

```python
IQR = Q3 - Q1
```

##‡πÅ‡∏™‡∏î‡∏á‡∏Ñ‡πà‡∏≤‡∏ä‡πà‡∏ß‡∏á‡∏Ñ‡∏ß‡∏≠‡πÑ‡∏ó‡∏•‡πå (Interquartile Range) ‡∏Ç‡∏≠‡∏á‡∏≠‡∏≤‡∏¢‡∏∏
*‡∏ú‡∏•‡∏•‡∏±‡∏û‡∏ò‡πå‡∏à‡∏∞‡πÑ‡∏î‡πâ 2.0 ‡∏Ñ‡∏∑‡∏≠‡∏Ñ‡πà‡∏≤‡∏ä‡πà‡∏ß‡∏á‡∏Ñ‡∏ß‡∏≠‡πÑ‡∏ó‡∏•‡πå‡∏´‡πà‡∏≤‡∏á‡∏Å‡∏±‡∏ô‡πÅ‡∏Ñ‡πà 2 ‡∏ä‡πà‡∏ß‡∏á

```python
IQR
```

##‡∏Ñ‡πà‡∏≤‡∏Ç‡∏≠‡∏ö‡∏•‡πà‡∏≤‡∏á‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö‡∏ï‡∏£‡∏ß‡∏à‡∏à‡∏±‡∏ö‡∏Ñ‡πà‡∏≤‡∏ú‡∏¥‡∏î‡∏õ‡∏Å‡∏ï‡∏¥ (outlier)
*lower_bound = Q1 - 1.5 * IQR  
##‡∏Ñ‡πà‡∏≤‡∏Ç‡∏≠‡∏ö‡∏ö‡∏ô‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö‡∏ï‡∏£‡∏ß‡∏à‡∏à‡∏±‡∏ö‡∏Ñ‡πà‡∏≤‡∏ú‡∏¥‡∏î‡∏õ‡∏Å‡∏ï‡∏¥ (outlier)
*upper_bound = Q3 + 1.5 * IQR

```python
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
```

##‡πÉ‡∏´‡πâ‡πÅ‡∏™‡∏î‡∏á‡∏ú‡∏•‡∏Ç‡∏≠‡∏á ‡∏Ñ‡πà‡∏≤‡∏Ç‡∏≠‡∏ö‡∏•‡πà‡∏≤‡∏á‡πÅ‡∏•‡∏∞‡∏Ñ‡πà‡∏≤‡∏Ç‡∏≠‡∏ö‡∏ö‡∏ô‡∏Ç‡∏≠‡∏á‡∏≠‡∏≤‡∏¢‡∏∏

```python
print(lower_bound)
print(upper_bound)
```

```python
df_no_outliers = df[(df['‡∏≠‡∏≤‡∏¢‡∏∏'] >= lower_bound) & (df['‡∏≠‡∏≤‡∏¢‡∏∏'] <= upper_bound)]
```
